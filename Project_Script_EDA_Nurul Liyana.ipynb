```python
price = list(df1['Motor Price'])
name = list(df1['Motor name'])

for value in range(len(price)):
    if '-' in price[value]:
        price[value] = price[value].replace('-','0')
    elif '1166612666' in price[value]:
        price[value] = price[value].replace('1166612666','12666')
    if 'Honda CRF450' in name[value]:
        price[value] = 13860
    if 'Honda CB650' in name[value]:
        price[value] = 25000
    if name [value] == 'Piaggio Liberty 125':
        price[value] = 3900
    if name [value] == 'Suzuki DR650SE':
        price[value] = 9900
    if 'Suzuki SV650' in name[value]:
        price[value] = 11820
    if name[value] == 'SYM Jet 4 125':
        price[value] = 5500
    if name[value] == 'Yamaha Zuma 125':
        price[value] = 3700
    if name[value] == 'Yamaha XT1200Z Super Tenere':
        price[value] = 16800
```
