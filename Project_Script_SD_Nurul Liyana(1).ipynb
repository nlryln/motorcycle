Basic Functions for easy understanding what is happening in the loop, and will be needing it few times in this notebook.

```python
def scroll_to_end():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

def next_page():
    button = driver.find_element(By.LINK_TEXT,'next')
    button.click()
    time.sleep(3)

def go_to_page(clas):
    driver.get(clas)
    time.sleep(3)
```

- Scroll to the end of the page : scroll_to_end()

- Clicked the 'next page' button : next_page():

- Goes to the page that is given in the bracket() : go_to_page(clas):


A while loop to get the links from used bike, from page to page. The for loop is to get the link  in each tag name. I delete the first 19 rows and the last 25 rows as its not the bikes page link.

```python
count = 0
draft = []
go_to_page(used_bike)
while driver.find_element(By.LINK_TEXT,'last') != None:
    links = []
    scroll_to_end()
    profile = driver.find_elements(By.TAG_NAME,"a")

    for link in profile:
        href = link.get_attribute('href')
        links.append(href)

    del links[:19]
    del links[-25:]
    draft.append(links)
    count += 1

    try:
        next_page()
    except NoSuchElementException as exc:
        print(exc)
        print("Total page:",count)
        break
       
```

## BIKEMART.SG

Since I had no prices in the new bikes pages, I went to search for prices in Bikemart.sg

```python
source = requests.get('https://bikemart.sg/motorcycle-price-list/')
page = BeautifulSoup(source.content, "html.parser") 
```

For this website, I dont have to go page by page to extract, I could get all the items in one find, which was easier too. I only had to change the name of motorcycle names as it was it wasn't the same format as the website above. 

```python
motor_list = []
price_list = []
year_list = []
```

```python
motor = page.find_all(name='td')
for item in motor:
    motorprice = item.get_text().strip()

    if motorprice != '':
        if ' (Used)' in motorprice:
            motor_name = motorprice.replace(' (Used)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' New (50th Anniversary Edition)' in motorprice:
            motor_name = motorprice.replace(' New (50th Anniversary Edition)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' New (Special Edition)' in motorprice:
            motor_name = motorprice.replace(' New (Special Edition)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' New (Bikers)' in motorprice:
            motor_name = motorprice.replace(' New (Bikers)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' New (Ohlins)' in motorprice:
            motor_name = motorprice.replace(' New (Ohlins)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' (New) (Bikers & Ohlins)' in motorprice:
            motor_name = motorprice.replace(' (New) (Bikers & Ohlins)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' (New) (Bikers)' in motorprice:
            motor_name = motorprice.replace(' (New) (Bikers)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' (Ohlins)' in motorprice: 
            motor_name = motorprice.replace(' (New) (Ohlins)','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' New' in motorprice:
            motor_name = motorprice.replace(' New','')
            year = motor_name[-4:]
            year_list.append(year)
            name = motor_name.replace(year,'')
            motor_list.append(name)
        elif ' (New)' in motorprice:
            motor_name = motorprice.replace(' (New)','')
            Abs = motor_name[-4:]
            num = motor_name[-2:]
            if Abs == ' ABS':
                year = 2022
                year_list.append(year)
                name = motor_name
            elif Abs == '-ABS':
                year = 1978
                year_list.append(year)
                name = motor_name
            elif num == '14':
                year = 2014
                year_list.append(year)
                name = motor_name
            else:
                year = motor_name[-4:]
                year_list.append(year)
                name = motor_name.replace(year,'')
            motor_list.append(name)
        elif '$' in motorprice:
            price = motorprice.replace('$','').replace(',','').replace('.00','')
            if len(price) >= 8:
                space = price[-5:]
                if space[0] == '1':
                    back = price[-5:] 
                    final = price.replace(price,back)
                    price_list.append(final)
                else:
                    back = price[-4:] 
                    final = price.replace(price,back)
                    price_list.append(final)
            else:
                price_list.append(price)
```
